```python
import pandas as pd
import urllib.parse
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# 엑셀 파일 경로 설정
file_path = r'C:\Users\USER\Desktop\5_6\core_pj\data\환자식_imageO_siteO_108개.xlsx'

# 엑셀 파일 로드
df = pd.read_excel(file_path)

# '제조사명'이 '(주)현대그린푸드 스마트푸드센터'인 데이터 필터링
filtered_df = df[df['제조사명'] == '(주)현대그린푸드 스마트푸드센터']

# '식품명' 값들을 리스트로 추출 및 수정
product_names = []
for name in filtered_df['식품명']:
    original_name = name.strip()
    if '&' in name:
        name = name.split('&')[0]
    if '/' in name:
        name = name.split('/')[0]
    if '[' in name:
        name = name.split(' ')[1]
    search_name = name.strip()
    product_names.append((original_name, search_name))

# 검색 URL 생성 및 저장
base_url = "https://www.greating.co.kr/search/searchCareResult?IL_ITEM_TEXT="
search_urls = []

for original_name, search_name in product_names:
    search_query = urllib.parse.quote(search_name)
    search_url = f"{base_url}{search_query}#page1"
    search_urls.append((original_name, search_url))

# Selenium 설정
service = ChromeService(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 크롤링 결과 저장
results = []

# 각 검색 URL에 대해 크롤링 수행
for original_name, url in search_urls:
    driver.get(url)
    
    # 상품이 로드될 때까지 대기
    wait = WebDriverWait(driver, 5)
    try:
        # 상품 페이지의 URL 가져오기
        product_url = driver.current_url
        
        # 가격 정보 가져오기
        price_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#dvSrchItemList_1458 > li > span > span')))
        price_text = price_element.text
        
        # 정규식을 사용하여 숫자만 추출
        match = re.search(r'\d+(?:,\d+)*', price_text)
        price = match.group().replace(',', '') if match else '가격 정보 없음'

        # 이미지 URL 가져오기
        image_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#dvSrchItemList_1458 > li > div > a > figure > img')))
        image_url = image_element.get_attribute('src')
        
        # 결과 저장
        results.append({
            '식품명': original_name,
            '제조사명': '(주)현대그린푸드 스마트푸드센터',  # 추가된 칼럼
            '가격': price,
            'site_url': product_url,
            'image_url': image_url,
        })
    except Exception as e:
        print(f"Error processing URL {url}: {e}")
        continue

# 브라우저 닫기
driver.quit()

# 결과를 DataFrame으로 변환
df_results = pd.DataFrame(results)
print(df_results)

# 엑셀 파일로 저장
output_file_path = r'C:\Users\USER\Desktop\5_6\core_pj\data\(주)현대그린푸드 스마트푸드센터.xlsx'
df_results.to_excel(output_file_path, index=False)

print(f"Results saved to {output_file_path}")
```

## 건강한사람들 주식회사

```python
import pandas as pd
import urllib.parse
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# 엑셀 파일 경로 설정
file_path = r'C:\Users\USER\Desktop\5_6\core_pj\data\환자식_imageO_siteO_107개_0520_1625.xlsx'

# 엑셀 파일 로드
df = pd.read_excel(file_path)

# '제조사명'이 '건강한사람들 주식회사'인 데이터 필터링
filtered_df = df[df['제조사명'] == '건강한사람들 주식회사']

# '식품명' 값들을 리스트로 추출 및 수정
product_names = []
for name in filtered_df['식품명']:
    # 4번째 글자부터 가져오기
    search_name = name[3:]
    product_names.append((name.strip(), search_name.strip()))

# 검색 URL 생성 및 저장
base_url = "https://dm.pulmuone.com/product/productList.do?sw="
search_urls = []

for original_name, search_name in product_names:
    search_query = urllib.parse.quote(search_name)
    search_url = f"{base_url}{search_query}#page1"
    search_urls.append((original_name, search_url))

# Selenium 설정
service = ChromeService(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 크롤링 결과 저장
results = []

# 각 검색 URL에 대해 크롤링 수행
for original_name, url in search_urls:
    driver.get(url)
    
    # 상품이 로드될 때까지 대기
    wait = WebDriverWait(driver, 10)
    try:
        # 상품 클릭
        product_element = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#page-contents > section > ul > li > figure > figcaption > a > dl')))
        product_element.click()
        
        # 상품 페이지의 URL 가져오기
        product_url = driver.current_url
        
        # 가격 정보 가져오기
        price_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#product_info > figcaption > dl > dd > div.price > div')))
        price_text = price_element.text
        
        # 정규식을 사용하여 숫자만 추출
        match = re.search(r'\d+(?:,\d+)*', price_text)
        price = match.group().replace(',', '') if match else '가격 정보 없음'

        # 이미지 URL 가져오기
        image_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#product_info > div > div > div > div > div > img')))
        image_url = image_element.get_attribute('src')
        
        # 결과 저장
        results.append({
            '식품명': original_name,
            '가격': price,
            'site_url': product_url,
            'image_url': image_url,
            '제조사명': '건강한사람들 주식회사'  # 추가된 칼럼
        })
    except Exception as e:
        print(f"Error processing URL {url}: {e}")
        continue

# 브라우저 닫기
driver.quit()

# 결과를 DataFrame으로 변환
df_results = pd.DataFrame(results)
print(df_results)

# 엑셀 파일로 저장
output_file_path = r'C:\Users\USER\Desktop\5_6\core_pj\data\건강한사람들 주식회사.xlsx'
df_results.to_excel(output_file_path, index=False)

print(f"Results saved to {output_file_path}")
```
